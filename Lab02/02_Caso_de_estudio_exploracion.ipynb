#### Determina la cantidad de películas que fueron producidas cada año, de 2006 a 2016

```python
peliculas_producidas = df.query('title_year >= 2006 & title_year <= 2016')
#### Grafica la cantidad de películas producidas cada año, de 2006 a 2016
peliculas_producidas.title_year.hist(figsize=(15,15), bins=30)
plt.xlabel("Year")
plt.ylabel("Counts")
plt.title("Movies produced per year")
```

```python
# grouped_director_earnings = df.groupby('director_name')[['gross']].sum().sort_values(by='gross', ascending=False)
grouped_director_earnings = df.groupby('director_name')['gross'].aggregate('sum').sort_values(ascending=False).reset_index()

top_twenty_earning_directors = grouped_director_earnings.head(20).director_name

top_twenty_earning_directors
```

### ¿Cuáles son las películas del género Sci-Fi de más bajo presupuesto


```python
scifi_movies = df[df.genres.str.contains('Sci-Fi')].sort_values(by="budget")
scifi_movies[['movie_title', 'genres', 'budget']].head()
```

### ¿Cuáles son los actores que en promedio han participado en las películas más largas?


```python
actors_1_duration = df[['actor_1_name', 'duration']].rename(columns={'actor_1_name': 'actor', 'duration': 'duration_1'}).groupby('actor').duration_1.mean().reset_index()
actors_2_duration = df[['actor_2_name', 'duration']].rename(columns={'actor_2_name': 'actor', 'duration': 'duration_2'}).groupby('actor').duration_2.mean().reset_index()
actors_3_duration = df[['actor_3_name', 'duration']].rename(columns={'actor_3_name': 'actor', 'duration': 'duration_3'}).groupby('actor').duration_3.mean().reset_index()

print(actors_1_duration.head())
print(actors_2_duration.head())
print(actors_3_duration.head())
```

```python
actors_duration = actors_1_duration.merge(actors_2_duration, on='actor').merge(actors_3_duration, on='actor')
actors_duration.head()
```

```python
actors_duration['total_avg_duration'] = actors_duration.apply(lambda row: np.mean([row.duration_1, row.duration_2, row.duration_3]), axis=1)
actors_duration.sort_values(by='total_avg_duration', ascending=False).head()
```
